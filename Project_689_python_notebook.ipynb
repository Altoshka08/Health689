#Creating SCD1 Dimension and Key maintenance

```python
#Creating a patient_dim
DimPatient = metabric_df[['patient_id', 'age_at_diagnosis', 'death_from_cancer']].copy()

DimPatient = pd.DataFrame(DimPatient)

DimPatient.head(10)
```

```python
#Load the data frames into the database tables
import psycopg2
from sqlalchemy import create_engine, types
db_params = {
    'database': 'Project',
    'user': 'postgres',
    'password': '0880',
    'host': 'localhost',
    'port': '5432',
}

conn = psycopg2.connect(**db_params)

engine = create_engine(f'postgresql+psycopg2://{db_params["user"]}:{db_params["password"]}@{db_params["host"]}:{db_params["port"]}/{db_params["database"]}')

dtype = {
    'patient_id': types.Integer,
    'age_at_diagnosis': types.Float,
    'death_from_cancer': types.String(length=25)
}

# Set the primary key constraint while loading DataFrame into SQL table
DimPatient.to_sql('patient_dim', engine, if_exists='replace', index=False, dtype=dtype)

# Execute SQL command to set primary key
with engine.connect() as con:
    con.execute('ALTER TABLE Patient_Dim ADD PRIMARY KEY (patient_id);')
```

```python
# Executing SCD type1 in Python

updating_data = {
    'patient_id': [5],
    'age_at_diagnosis': [87.3],  # Updated ages
    'death_from_cancer': ['Died of Disease']  # Updated death status
}
updating_dataset = pd.DataFrame(updating_data)

for index, row in updating_dataset.iterrows():
    patient_id = row['patient_id']
    age = row['age_at_diagnosis']
    death = row['death_from_cancer']
    
    update_query = f"""
    UPDATE patient_dim
    SET age_at_diagnosis = {age}, death_from_cancer = '{death}'
    WHERE patient_id = {patient_id}
    """
    
    # Execute the update query
    with engine.connect() as con:
        con.execute(update_query)
```

```python
Drug_Dim = drugs_df[['Name', 'Category', 'Primary Site' ]].copy()
Drug_Dim.rename(columns = {'Name' : 'drug_name', 'Category':'drug_category','Primary Site': 'drug_primary_site'}, inplace=True)
Drug_Dim.reset_index(drop=True, inplace=True)
Drug_Dim.index += 1  # Shift the index by 1
Drug_Dim.rename_axis('drug_id', inplace=True)

Drug_Dim.head(10)
```

```python
Drug_Dim.to_sql('drug_dim', engine, if_exists='append', index=False, dtype=dtype)
conn.commit()
conn.rollback()
```
